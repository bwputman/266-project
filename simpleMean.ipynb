```python
# The most basic model we can think of in 5 lines of code, a simple mean, achieves a Kaggle score of 0.55411
# Any additional complexity will need to yield significant profit

import pandas as pd
from IPython.display import Image

questionpairs_train = pd.read_csv('/home/geoffrey_link/train.csv')
questionpairs_test  = pd.read_csv('/home/geoffrey_link/test.csv')

simple_probability  = questionpairs_train['is_duplicate'].mean()

naive_submission    = pd.DataFrame({
                                    'test_id'      : questionpairs_test['test_id'], 
                                    'is_duplicate' : simple_probability
                                  })

naive_submission.to_csv('/home/geoffrey_link/simpleMean_submission.csv', index=False)
print naive_submission.head()
Image(filename='simpleMean_submission.png') 
```
